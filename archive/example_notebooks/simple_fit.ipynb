```python
def get_data(n_timepoints=100, n_channels=10):
    import numpy as np
    import pandas as pd
    return pd.DataFrame(np.random.rand(n_timepoints, n_channels))

def get_paradigm(n_timepoints=100):
    # All between 0 and 180 degrees in radians
    import numpy as np
    import pandas as pd
    return pd.DataFrame(np.random.rand(n_timepoints) * np.pi)
```

```python
from braincoder.stimuli import OneDimensionalRadialStimulus
```

```python
from braincoder.models import GaussianPRF, VonMisesPRF
from braincoder.optimize import ParameterFitter
import numpy as np

data = get_data() # Neural data n_timepoints x n_channels
paradigm = get_paradigm() # Orientations
model = VonMisesPRF(paradigm=paradigm, data=data)

mus = np.linspace(0, np.pi, 8)
sds = np.array([0.1, 0.5, 1.0])
amplitudes = [1.0]
baselines = [0.0]

fitter = ParameterFitter(model=model, data=data, paradigm=paradigm)

# Do a quick grid search first
grid_pars = fitter.fit_grid(mus, sds, amplitudes, baselines, use_correlation_cost=True)

# Optimize amplitude and baseline using OLS
grid_pars = fitter.refine_baseline_and_amplitude(grid_pars)

# Use gradient descent to further refine mu and sd
gd_pars = fitter.fit(init_pars=grid_pars)
r2 = fitter.get_rsq(gd_pars)
```
