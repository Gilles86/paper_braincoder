```python
from nilearn import input_data
import numpy as np
from braincoder.models import GaussianPRF2D
from braincoder.optimize import ParameterFitter

masker = input_data.NiftiMasker('/data/brain_mask.nii.gz')
data = '/data/average_timesries.nii.gz'

data = masker.fit_transform(data)
paradigm = np.load('/data/paradigm.npy')
grid_coords = np.load('/data/grid_coords.npy')

model = GaussianPRF2D(data=data, paradigm=paradigm, grid_coords=grid_coords,)

fitter = ParameterFitter(model=model, paradigm=paradigm) 

mus_x = np.linspace(-10, 10, 20)
mus_y = np.linspace(-10, 10, 20)
sds = np.linspace(0.1, 10, 20)
amplitudes = [1.0]
baselines = [0.0]

grid_pars = fitter.fit_grid(mus_x, mus_y, sds, amplitudes, baselines, use_correlation_cost=True)
grid_pars = fitter.refine_baseline_and_amplitude(grid_pars)
gd_pars = fitter.fit(learning_rate=0.01)
```

```python
# Initialize model
model = GaussianPRF2D(data=data, paradigm=paradigm, grid_coords=grid_coords, parameters=gd_pars)

# predict
predictions = model.predict()

# simulate with Gaussian noise
simulated_data = model.simulate()

# Simulate with non-spherical noise
sigma = np.full((gd_pars.shape[1], gd_pars.shape[1]), 0.3)  # Fill the matrix with 0.3
np.fill_diagonal(sigma, 1.0)  # Set the diagonal elements to 1.0

simulated_data_non_spherical = model.simulate(sigma=sigma)
```
